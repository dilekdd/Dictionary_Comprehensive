Create a dictionary where the keys are numbers and the values are their squares

```python
squares = {x: x**2 for x in range(10)}
print(squares)
```

Get the squares of the values in the given dictionary

```python
dictionary = {'a': 1,
             'b': 2,
             'c': 3,
             'd': 4}
{k: v ** 2 for (k, v) in dictionary.items()}
```

Convert the keys of the given dictionary to uppercase

```python
dictionary = {'a': 1,
             'b': 2,
             'c': 3,
             'd': 4}
{k.upper(): v for k, v in dictionary.items()}
```

Convert the keys of the given dictionary to uppercase and multiply the values by 2

```python
dictionary = {'a': 1,
             'b': 2,
             'c': 3,
             'd': 4}

{k.upper(): v * 2 for k, v in dictionary.items()}
```

Create a dictionary with the even numbers between 0 and 10. Keys should be the even numbers in the range 0-10, Values should be the squares of those numbers.

```python
numbers = range(10)
{n: n ** 2 for n in numbers if n % 2 == 0}
```

Create a dictionary with the numbers between 0 and 10. Keys should be the numbers in the range, Values should be the squares of those numbers if the number is en even number, if no, multiply the value with 3

```python
numbers = range(10)
{n: (n ** 2 if n % 2 == 0 else n * 3) for n in numbers}
```

Create a dictionary that shows the prime numbers in the range of 0 - 12 as keys and the Values should be shown as TRUE or FALSE

```python
def prime_check(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5) +1):
        if n % i == 0:
            return False
    return True

prime_dictionary = {n: prime_check(n) for n in range(2, 12)}
print(prime_dictionary)
            
```

Create a dictionary comprehension that maps each prime number between 2 and 22 to its square

```python
def prime_square(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) +1):
        if n % i == 0:
            return False
    return True

dictionary_new = {n: n ** 2 for n in range(2, 22) if prime_square(n)}
print(dictionary_new)
```
